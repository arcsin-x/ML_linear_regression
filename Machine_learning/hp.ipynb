```python
from sklearn.cross_validation import train_test_split
target = data_train['price']
train = data_train.drop('price',axis = 1)
x_train, x_test, y_train, y_test = train_test_split(train, target, test_size=0.1, random_state=2)
```

Используем регрессию градиентного спуска. Хорошо применим к смеси категориальных и непрерывных признаков, не требует масштабирования. Данные не разрежены. 
Здесь работает лучше, чем линейная регрессия.
Вообще я решила использовать метод градиентного спуска потому что в данном случае могу себе это позволить :)

```python
from sklearn import ensemble
grad = ensemble.GradientBoostingRegressor(n_estimators=700, max_depth=3, min_samples_split=2, loss='ls')
grad.fit(x_train, y_train)
```

```python
print(grad)
```

```python
from math import sqrt
rmse1 = (sqrt(grad.score(x_train,y_train)))
rmse2 = (sqrt(grad.score(x_test,y_test)))
```

```python
print ("Правильность на обучающем наборе: {:.2f}".format(rmse1))
print ("Правильность на тестовом наборе: {:.2f}".format(rmse2))
```
